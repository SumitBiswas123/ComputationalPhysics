# Euler integration
In this notebook we use the Euler method to integrate a simple first order differential equation.
For this, we have
$$
  \frac{d S(t)}{dt} = F(t, S(t)),
$$
where F(t) is a function that returns the time derivative of S(t). 
For our code we will choose:
$$
F(t) = \exp(-t),
$$
and the initial condition 
$$
S(t=0) = -1.
$$

We then have that the solution can be found analytically as:
$$
S(t) = -\exp(-t).
$$

```python
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-poster')
%matplotlib inline

# define the parameters
f = lambda t, s: np.exp(-t) # ODE
h = 0.1                     # the step size
t = np.arange(0,1+h, h)     # the numerical grid
s0 = -1.0                   # the initial condition
```

```python
# Explicit Euler method
s = np.zeros(len(t))        # create a vector the size of t, all values are zero
s[0] = s0                   # the initial value of s is s0

for i in range(0, len(t)-1):  
    s[i+1] = s[i] + h*f(t[i],s[i])
```

```python
# Plotting our results
plt.figure(figsize = (12, 8))
plt.plot(t, s, 'ro--', label='Approximate')      # the solution from applying the Euler method
plt.plot(t, -np.exp(-t), 'g', label='Exact')     # the analytic solution
plt.title('Approximate and Exact Solution \
for Simple ODE')
plt.xlabel('t')
plt.ylabel('f(t)')
plt.grid()
plt.legend(loc='lower right')
plt.show()
```
